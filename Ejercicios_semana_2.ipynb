Vamos a crear un programa que lea los resultados de los partidos de la liga española en el año 2019-2020, y nos devuelva información sobre estos datos. El fichero está subido junto con los ejercicios.

El programa ofrecerá un menú, para seleccionar la información deseada:

**Estadística de un equipo**: Nos pide por teclado el nombre de un equipo y nos muestra el número de goles que ha metido, los paridos ganados, perdidos y empatados.

**Nombres de equipos**: Nos muestra la lista de equipos que juegan.

**Clasificación de la liga**: Nos muestra los tres primeros equipos de la liga.

**Quiniela por fecha:** Introducimos una fecha y nos dice los resultados de la quiniela de ese día.

Salir


-----------------------------


Para realizar este programa podemos realizar las siguientes funciones:

**menu()**: Muestra el menú y devuelve un entero con la opción escogida.

**LeerPartidos()**: Función que lee el fichero y devuelve una lista con los partidos (cada partido se va a guardar en un diccionario).

**SumarGoles(equipo)**: Función que recibe un nombre de un equipo y devuelve el total de goles metidos.

**InfoEquipos(equipo)**: Función que recibe un nombre de un equipo y devuelve una lista con los paridos ganados, perdidos y empatados.

**Equipos()**: Función que devuelve una lista con todos los equipos.

**Quiniela(dia,mes,año)**: Función que recibe el día, el mes y el año. Y devuelve una lista con los partidos y resultados de la quiniela.

````
import datetime
from tabulate import tabulate

print("****************************************")
print("** Estadistica de la liga española")
print("****************************************")
listaPartidos =[];

def menu():
  opcionSeleccionada =input("Seleccione una opción:\n\t(1) Estadistica de un equipo\n\t(2) NOmbre de equipos\n\t(3) Clasificacion de la liga\n\t(4) Quiniela por fecha\n\t(5) Salir \n")
  
  while opcionSeleccionada not in ["1", "2", "3", "4", "5"]:
    print("La opcion ingresada no es valida")
    opcionSeleccionada =input("Seleccione una opción:\n\t(1) Estadistica de un equipo\n\t(2) NOmbre de equipos\n\t(3) Clasificacion de la liga\n\t(4) Quiniela por fecha\n\t(5) Salir ")

  return opcionSeleccionada

def leerPartidos():
  
  with open('la-liga-2019.csv') as _file:
    fichero = _file.readlines()
    listaPartidos =[];
    
    for i in fichero:
      linea = i.split(",")
      partido = {}
      if linea[0] !="Round Number":
        partido["Round_Number"] = linea[0]
        partido["Date"] = linea[1]
        partido["Location"] = linea[2]
        partido["HomeTeam"] = linea[3]
        partido["AwayTeam"] = linea[4]
        partido["Result"] = linea[5]
        resultado = linea[5].split("-")
        #print(resultado)
        partido["HomeTeamResult"] = int(resultado[0])
        partido["AwayTeamResult"] = int(resultado[1])
        listaPartidos.append(partido)
  
  return listaPartidos    

def sumarGoles(equipo):

  acumuladorGoles = 0
  datos = leerPartidos()

  for dicc in datos:
    #print(dicc)
    for clave in dicc:
      #print(clave)
      if clave == 'HomeTeam' and dicc[clave] == equipo:
        acumuladorGoles = acumuladorGoles + int(dicc["HomeTeamResult"])
      
      if clave == 'AwayTeam' and dicc[clave] == equipo:
        acumuladorGoles += dicc["AwayTeamResult"]
  return acumuladorGoles

def InfoEquipos(equipo):

  datos = leerPartidos()
  estadisticaEquipo = []
  listaPartidoGanados = []
  listaPartidosPerdidos = [] 

  for dicc in datos:
    for clave in dicc:
      if ((clave == 'HomeTeam' and dicc[clave] == equipo ) and (int(dicc["HomeTeamResult"]) > int(dicc["AwayTeamResult"]))) or ((clave == 'AwayTeam' and dicc[clave] == equipo ) and (int(dicc["HomeTeamResult"]) < int(dicc["AwayTeamResult"]))):
         listaPartidoGanados.append(dicc)
      
      if ((clave == 'HomeTeam' and dicc[clave] == equipo ) and (int(dicc["HomeTeamResult"])< int(dicc["AwayTeamResult"]))) or ((clave == 'AwayTeam' and dicc[clave] == equipo ) and (int(dicc["HomeTeamResult"]) > int(dicc["AwayTeamResult"]))):
         listaPartidosPerdidos.append(dicc)

  estadisticaDicc = {}
  estadisticaDicc["partidosGanados"] = listaPartidoGanados
  estadisticaDicc["partidosPerdidos"] = listaPartidosPerdidos
  estadisticaEquipo.append(estadisticaDicc)
  return estadisticaEquipo

def Equipos():
  datos = leerPartidos()
  listaEquipos = []
  for i in datos:
    if i['HomeTeam'] not in listaEquipos:
      listaEquipos.append(i['HomeTeam'])
    
    if i['AwayTeam'] not in listaEquipos:
      listaEquipos.append(i['AwayTeam'])
  return listaEquipos

def Quiniela(dia,mes,año):
  try:
    datos = leerPartidos()
    listaPartidos = []
    for i in datos:
      fecha_str = i['Date']
      fecha_Jornada_obj = datetime.datetime.strptime(fecha_str, '%d/%m/%Y %H:%M')

      fecha_Quinela = dia+"/"+mes+"/"+año
      fecha_Quinela_obj = datetime.datetime.strptime(fecha_Quinela, '%d/%m/%Y')
      if fecha_Jornada_obj.date() == fecha_Quinela_obj.date():
        listaPartidos.append(i)
    return listaPartidos
  except:
    print("La fecha ingresada no es valida")

def tablaPosiciones():
  listaEquipos = Equipos()
  listaClasificacion = []
  for i in listaEquipos:

    puntos = 0
    listaPartidos = InfoEquipos(i)
    for item in listaPartidos:
      listaPartidosGanados = item["partidosGanados"]
      puntos = len(listaPartidosGanados)*3
      
      dic = {}
      dic["equipo"] = i
      dic["puntos"] = puntos
      listaClasificacion.append(dic)
  
  return listaClasificacion  


opcion = menu()
encabezado = ['Round Number','Date','Location','Home Team','Away Team','Result']
def obtenerLinea(dic):
  linea =[]
  linea.append(dic["Round_Number"])
  linea.append(dic["Date"])
  linea.append(dic["Location"])
  linea.append(dic["HomeTeam"])
  linea.append(dic["AwayTeam"])
  linea.append(dic["Result"])
  return linea

if opcion == "1":
  try:
    
    equipo = input("Ingrese el nombre de equipo: ")
    print("==============================================")
    print("Goles metidos: ", sumarGoles(equipo))
    listaPartidos = InfoEquipos(equipo)
      
    for i in listaPartidos:
      listaPartidosGanador = i["partidosGanados"]
      listaPartidosPedidos = i["partidosPerdidos"]
      
      listaLineas = []
      for dic in listaPartidosGanador:
        linea = obtenerLinea(dic)
        listaLineas.append(linea)
      print("Partidos ganados:")
      print(tabulate( listaLineas , headers=encabezado))

      listaLineas = []
      for dic in listaPartidosPedidos:
        linea = obtenerLinea(dic)
        listaLineas.append(linea)
      print("Partidos perdidos:")
      print(tabulate( listaLineas , headers=encabezado))

  except:
    print("El equipo ingresado no existe")
elif opcion == "2":

  print("Lista de equipos")
  print("==========================")
  lista = Equipos()
  lista.sort()
  for i in lista:
    print(i)
elif opcion == "3":

  print("Clasificacion de equipos")
  print("==========================")
  lis =  tablaPosiciones()
  newList = sorted(lis, key = lambda i: i['puntos'], reverse=True)
  firtsTree = newList[:3]
  #print(firtsTree)
  encabezado = ["Equipo", "Ptos"]
  detalle = []
  for i in firtsTree:
    linea = []
    linea.append(i["equipo"])
    linea.append(i["puntos"])

    detalle.append(linea)
  print(tabulate( detalle , headers=encabezado))

elif opcion == "4":
  dia = input("Introduzca dia(ejemplo: 01):")
  mes = input("Introduzca mes(ejemplo: 06):")
  anno = input("Introduzca año(ejemplo: 2021):")

  datos = Quiniela(dia, mes, anno)
  listaLineas = []
  for dic in datos:
    linea = obtenerLinea(dic)
    listaLineas.append(linea)
  print("Paridos de la jornada")
  print(tabulate( listaLineas , headers=encabezado))





````
